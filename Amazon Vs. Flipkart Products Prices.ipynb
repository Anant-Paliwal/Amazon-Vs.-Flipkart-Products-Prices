# Now lets merge both the Dataset and compare both the biggest E-commerce sites

```python
flipkart.head(1)
```

```python
amazon.head(1)
```

```python
joint_data=amazon.merge(flipkart,left_on="product_name in Amazon",right_on='product_name in Flipkart',how='inner')
```

```python
joint_data.head()
```

# Lets Delete the duplicate records

```python
joint_data_1=joint_data.drop_duplicates()
joint_data_1
```

# lets check percentage of null values

```python
(joint_data_1.isnull().sum()/joint_data_1.shape[0])*100
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(joint_data_1.isnull(),cbar=False,yticklabels=False)
```

# Now as our Dataset is clean Lets analyze its

```python
joint_data_1.describe()
```

# Amazon Prices Vs. Flipkart Prices

```python
plt.figure(figsize=(15,8))
sns.set(style="darkgrid")
m = plt.hist(joint_data_1["retail_price in Amazon"],bins=180,fc = (0,0,1,0.9),label="Amazon Prices")
b = plt.hist(joint_data_1["retail_price in Flipkart"],bins=180,fc = (0,1,0,0.6),label = "Flipkart Prices")
plt.legend()
plt.xlabel("Range of Prices")
plt.ylabel("Frequency")
plt.title("Amazon Prices Vs. Flipkart Prices")
```

# From the above visuals we can conclude that the prices of Product on Flipkart are slightly higher than that on Amazon

# Amazon Discount Vs. Flipkart Discount

```python
plt.figure(figsize=(15,8))
sns.set(style="darkgrid")
m = plt.hist(joint_data_1["discounted_price in Amazon"],bins=60,fc = (0,0,1,0.7),label="Amazon Discounted")
b = plt.hist(joint_data_1["discounted_price in Flipkart"],bins=60,fc = (0,1,0,0.6),label = "Flipkart Discounted")
plt.legend()
plt.xlabel("Range of Discounted")
plt.ylabel("Frequency")
plt.title("Amazon Discounted Vs. Flipkart Discounted")
```

# Visuals above clearly states that Flipkart and Amazon same discounted most of the products

# Flipkart prices Vs. Flipkart Discounted

```python
sns.set(style="darkgrid")
sns.jointplot(x="discounted_price in Flipkart",y="retail_price in Flipkart",data=joint_data_1)
```

# Products Price between 200000-500000 are the mostly Discounted Products on Flipkart
